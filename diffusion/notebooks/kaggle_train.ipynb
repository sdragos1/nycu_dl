```python
import os
import subprocess

REPO_URL    = "https://github.com/sdragos1/nycu_dl.git"
REPO_BRANCH = "diffusion"
REPO_DIR    = "/kaggle/working/nycu_dl"

if not os.path.exists(REPO_DIR):
    subprocess.run([
        "git", "clone",
        "--branch", REPO_BRANCH,
        "--single-branch",
        REPO_URL, REPO_DIR,
    ], check=True)
    print("Repository cloned.")
else:
    subprocess.run(["git", "-C", REPO_DIR, "pull"], check=True)
    print("Repository updated.")

os.chdir(f"{REPO_DIR}/diffusion")
print(f"Working directory: {os.getcwd()}")
```

```python
import glob

candidates = glob.glob("/kaggle/input/**/objects.json", recursive=True)
if candidates:
    DATA_ROOT = os.path.dirname(candidates[0])
    print(f"Auto-detected dataset at: {DATA_ROOT}")
else:
    DATA_ROOT = "/kaggle/input/iclevr-dataset"
    print(f"Could not auto-detect. Using fallback: {DATA_ROOT}")
```

```python
WANDB_KEY = ""
if WANDB_KEY:
    os.environ["WANDB_API_KEY"] = WANDB_KEY
    DISABLE_WANDB = "false"
    print("W&B enabled.")
else:
    DISABLE_WANDB = "true"
    print("W&B disabled — set WANDB_KEY to enable.")
```

```python
SAVE_DIR   = "/kaggle/working/results/unet"
EPOCHS     = 100
BATCH_SIZE = 32
LR         = 2e-4

cmd = [
    "python", "-m", "src.train",
    f"train.data_root={DATA_ROOT}",
    f"train.save_dir={SAVE_DIR}",
    f"train.epochs={EPOCHS}",
    f"train.batch_size={BATCH_SIZE}",
    f"train.lr={LR}",
    f"train.disable_wandb={DISABLE_WANDB}",
]

print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)
```
